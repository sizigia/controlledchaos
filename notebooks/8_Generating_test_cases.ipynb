```python
for i in range(1, 4):
    files = os.listdir('t5/')
    new_folder = f"sample{i}"
    os.mkdir(new_folder)
    mvfiles = random.sample(files, 50*i)
    for file in mvfiles:
        shutil.copy(f"{curr_folder}/t5/{file}", new_folder)
```
